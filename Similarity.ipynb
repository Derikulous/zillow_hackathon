```python
import pandas
import numpy as np
import scipy

from itertools import islice

from zillow_hackathon.dataset import Neighborhood

import sklearn.preprocessing
import scipy.spatial.distance
```

```python
nb1 = Neighborhood.get_for_city_and_neighborhood('Seattle', 'Capitol Hill')
nb2 = Neighborhood.get_for_city_and_neighborhood('San Francisco', 'Russian Hill')
```

```python
def vectorize(n):
    return [
        n.median_age or 0,
        n.median_income or 0
    ]
```

```python
all_nbrs = list(Neighborhood.get_neighborhoods_in_city('Seattle')) + \
                list(Neighborhood.get_neighborhoods_in_city('San Francisco'))
    
all_vecs = []
for n in all_nbrs:
    #if n.median_age == None:
    #    continue
        
    v = vectorize(n)
    all_vecs.append(v)
    
all_vecs = np.array(all_vecs)
```

```python
min_max_scaler = sklearn.preprocessing.MinMaxScaler()
min_max_scaler.fit(all_vecs)
all_vecs = min_max_scaler.transform(all_vecs)

print all_vecs
```

```python
x = min_max_scaler.transform(vectorize(nb1))
```

```python
def find_top_similar(v1, where):
    res = []
    
    for i, v2 in enumerate(where):
        d = 1 - scipy.spatial.distance.cosine(v1, v2)
        res.append((i, d))
        
    return sorted(res, key=lambda p: [ -p[1] ])        

#x = all_vecs[100]

top_similar = find_top_similar(x, all_vecs)
for r in islice(top_similar, 10):
    id = r[0]
    d = r[1]
    
    print "#%d: %s: %f" % (id, all_nbrs[id].name, d)

print 
print x
print all_vecs[top_similar[0][0]]
print all_vecs[top_similar[1][0]]
```
